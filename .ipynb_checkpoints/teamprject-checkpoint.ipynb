#### 데이터 전처리 개요

**데이터 작성**

```python
data = {'2017':[1145373, 407415], 
       '2018':[1986511, 538892], 
       '2019':[2004643, 586467], 
       '2020':[1559524, 1113736], 
       '2021':[1629129, 964440]}
```

```python
budget= pd.DataFrame(data=data, index=['매출액', '손실액'])
budget
```

**서울교통공사 연간 매출액 & 손실액 시각화**

```python
budget.T.plot()
plt.title('서울교통공사 연간 매출액 & 손실액')
plt.ylabel('단위: 0.4=4천억 / 2.0=2조')
budget
```

**서울교통공사 연간 매출액 &손실액 시각화 분석**
- 의도 : 서울교통공사 손실액의 추세를 파악해보기 위함이다. 

- 해석 : 매출액은 감소하고 있지만, 손실액은 증가해왔다. 특히 2020년도의 손실율이 가장 높았다.

- 결과 : 재정적으로 안좋은 상황이 지속되고 있는 것이 사실임을 확인했다.


**연간 매출액 대비 손실액 비율 구하기**

**손실액 비율(단위 : %) : 손실액 / 매출액 *100**

```python
lratio = round(budget.iloc[1] / budget.iloc[0]*100, 2)#round 소수점 두번째 수까지 표현
lratio
```

```python
budget2= budget.append(lratio, ignore_index=True)
budget2
```

```python
budget3= budget2.rename(index={0:'매출액', 1:'손실액', 2:'손실비율'})
budget3.index.name='서울교통공사'
budget3
```

#### 손실비율에 대한 증감율 =  [(최종수치-최초수치)/최초수치]*100
**ex**
- 2018년도 손실비율에 대한 증감율 (단위 : %)  = (2018손실비율-2017손실비율)/(2017 손실비율)*100

```python
#round 함수로 소수점 두번째 자리만 추출
hl18=round((budget3['2018']['손실비율']-budget3['2017']['손실비율'])/budget3['2017']['손실비율']*100, 2)
hl19=round((budget3['2019']['손실비율']-budget3['2018']['손실비율'])/budget3['2018']['손실비율']*100, 2)
hl20=round((budget3['2020']['손실비율']-budget3['2019']['손실비율'])/budget3['2019']['손실비율']*100, 2)
hl21=round((budget3['2021']['손실비율']-budget3['2020']['손실비율'])/budget3['2020']['손실비율']*100, 2)
```

```python
budget3.loc['손실비율에 대한 증감율']=[np.nan, hl18, hl19, hl20, hl21]
budget3
```

**결측치 제거(0으로 치환)**

```python
budget4=budget3.fillna(0)
budget4
```

```python
budget5=budget4.T
budget5
```

```python
budget6=budget5[['손실비율에 대한 증감율']]
budget6
```

**서울교통공사 연간 손실비율에 대한 증감율 시각화**

```python

budget6.plot()
plt.title('서울교통공사 연간 손실비율에 대한 증감율')
plt.ylabel('단위: %')
print("손실비율에 대한 증감율 =  [(최종수치-최초수치)/최초수치]*100")
print('ex =  2018년도 손실 손실비율에 대한 증감율 (단위 : %)  = (2018손실비율-2017손실비율)/(2017 손실비율)*100)')
budget6
```

```python
budget7=budget5[['손실비율']]
budget7
```

**서울교통공사 매출액 대비 손실비용 비율 시각화**

```python

budget7.plot()
plt.title('서울교통공사 매출액 대비 손실비용 비율')
plt.ylabel('단위: %')
print('손실액 비율(단위 : %) : 손실액 / 매출액 *100')
budget7
```

```python
budget8=budget5[['손실비율','손실비율에 대한 증감율']]
budget8.columns.name='연도'
budget8
```

**서울교통공사 매출액 대비 손실액 비율&손실비용 증감율 비교**

```python
budget9=budget8.T

budget8.plot()
plt.title('서울교통공사 손실율 & 손실율에 대한 증감율')
budget9
```

**서울교통공사 매출액 대비 손실액 비율&손실비용 증감율 비교 시각화 분석**

- 의도 : 해당 연도의 매출액이 커서 적자비용이 큰 것이 아닌지 오류를 범하지 않기 위해서 매출액 대비 손실액 즉 손실액의 상대성을 연도별로 확인

- 해석 : 매출액 내 손실액의 비율은 계속해서 증가해왔음을 알 수 있고, 손실액의 비율의 증감율을 따져보았을 때도 주황색선이 전반적으로 0% 이상으로 위치하는 것으로 보아 손실율은 또한 계속해서 증가해왔음을 알 수 있다.

**서울시 총인구& 서울시 거주 65세 고령자 인구 추이 시각화**
- 의도 : 서울시 총 인구수의 변화 추이와 서울시 65세 이상 고령자수의 변화 추이를 비교하기 위함이다.

- 해석 : 서울시 총 인구수는 상승하지 않았으나, 서울시 65세 이상 고령자수는 서서히 상승하고 있다.


**총인구수 대비 65세 고령자의 비율**

**총인구수 대비 65세 고령자의 비율 (단위 :%) : 65세이상 고령자 / 서울시 인구수 *100**

```python
seoul_count4=seoul_count3.loc['2017':'2021',:]#2017~2021년도 의 자료를 추출 (서울시 교통공사의 출범이 2017년도)
seoul_count4
seoul_count5=seoul_count4.astype(int)#자료의 값이 object이기떄문에 int형식으로 astype()함수로 변경
seoul_count5
```

```python
#round()함수로 소수 두번째 자리까지 출력
seoul_count5['비율']=round(seoul_count5['65세이상 고령자']/seoul_count5['서울시 인구수']*100,2)
seoul_count6=seoul_count5.drop(['서울시 인구수','65세이상 고령자'],axis = 1)#서울시 인구수 65세이상 고령자 columns 제거

seoul_count6
```

**총인구수 대비 65세 고령자의 비율 시각화**

```python
bar=seoul_count6.plot.bar()

for p in bar.patches:
    left, bottom, width, height = p.get_bbox().bounds
    bar.annotate(height, (left+width/2, height*1.01), ha='center')
plt.title('서울시 총인구수 대비 서울시 거주 65세 이상 노령자 수 비율(2017~2021)')
plt.ylabel('단위: %')
plt.ylim(8,20)
print('총인구수 대비 65세 고령자의 비율 (단위 :%) : 65세이상 고령자 / 서울시 인구수 *100')
seoul_count6
```

**서울시 총인구수 대비 서울시 거주 65세 이상 노령자 수 비율 시각화 분석**
- 의도 : 서울시 총 인구수 내 서울시 65세 이상 고령자수의 비율의 상승폭을 좀 더 자세히 파악하기 위함이다. “서울교통공사의 지하철 총 무임승차비용 내 노인 무임승차비용의 비율 분석”(3번)그래프의 비율 변화폭과 해당 그래프의 비율 변화폭을 비교하는 것이기에 2017년~2021년의 데이터를 추출한다.

- 해석 : 2017~2021 65세 이상 고령자수의 비율 변화폭은 크다.
